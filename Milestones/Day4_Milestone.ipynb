### How many different companies are represented in the data set?


```python
a1 = len(df['Company Name'].unique())
a1
```

### What is the total number of jobs created for businesses in Queens?


```python
df_Queens = df[df['Borough']=='Queens']
a2 = df_Queens['Job created'].sum()
a2
```

### How many different unique email domains names are there in the data set?

```python
emails = df['company email'].str.split('@', expand = True)
emails = emails.rename(columns = {0: 'email address', 1:'domain'})
a3 = len(emails['domain'].unique())-1 # subtract NaN
a3
```

### Considering only NTAs with at least 5 listed businesses, what is the average total savings and the total jobs created for each NTA?


```python
# group by NTAs
NTAs = df[['NTA', 'Company Name','Job created', 'Total Savings']].groupby(by='NTA')
```

```python
# number of companies in each NTA
NTAs['Company Name'].count()
```

```python
# NTAs with more than 5 companies
NTA5 = NTAs.filter(lambda x: x['Company Name'].count()>=5)
NTA5
```

```python
NTA5['Job created'].sum()
```

```python
NTA5['Total Savings'].mean()
```

```python
# I think this answers a different question, keep it for your own review but the answer is much more simple
NTA5[['NTA', 'Job created', 'Total Savings']].groupby(by='NTA').agg({'Job created' : 'sum',\
                                                                            'Total Savings' : 'mean'})
```
